**Important Equations**


Probability Oracle: 
$$
\begin{aligned}
O_{\mathrm{prob}}:\quad
&\sum_k a_k
|\theta_k\rangle_{\Theta}
|0\rangle_{\mathrm{prob}}
|0\rangle_{\mathrm{work}}
\\
&\longmapsto
\sum_k a_k
|\theta_k\rangle_{\Theta}
\left(
\sqrt{1-p(\theta_k)}\,|0\rangle_{\mathrm{prob}}
+
\sqrt{p(\theta_k)}\,|1\rangle_{\mathrm{prob}}
\right)
|\mathrm{work}(\theta_k;H_{\mathrm{prob}})\rangle_{\mathrm{work}} .
\end{aligned}
$$


$$
\begin{aligned}
H_{\mathrm{prob}}
&=
\sum_j c_j P_j,
\\
p(\theta_k)
&=
f\!\left(
\left\langle \psi(\theta_k)\left|
H_{\mathrm{prob}}
\right|\psi(\theta_k)\right\rangle
\right)
\\
&=
f\!\left(
\left\langle \psi(\theta_k)\left|
\sum_j c_jP_j
\right|\psi(\theta_k)\right\rangle
\right).
\end{aligned}
$$


Phase Oracle:  
$$

\sum_k a_k
|\theta_k\rangle_{\Theta}
|0\rangle_{\mathrm{prob}}
|0\rangle_{\mathrm{work}}
\longmapsto
\sum_k a_k
e^{-it
\left\langle \psi(\theta_k)\left|
\sum_j c_j P_j
\right|\psi(\theta_k)\right\rangle}
|\theta_k\rangle_{\Theta}
\left(
\sqrt{1-p(\theta_k)}\,|0\rangle_{\mathrm{prob}}
+
\sqrt{p(\theta_k)}\,|1\rangle_{\mathrm{prob}}
\right)
|\mathrm{work}(\theta_k;H_{\mathrm{prob}})\rangle_{\mathrm{work}}

$$
